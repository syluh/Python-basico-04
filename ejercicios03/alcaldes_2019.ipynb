```python
import pandas as pd
```

```python
datos_alcaldes = pd.read_spss('data/resultadosdignidadalcalde2019.sav')
```

```python
datos_alcaldes
```

```python
datos_candidatos = pd.read_spss('data/candidatos2019.sav')
datos_candidatos.shape
```

```python
datos_candidatos.head(15)
```

```python
datos_candidatos_2 = datos_candidatos[['CANDIDATO_CODIGO', 'CANDIDATO_NOMBRE', 'OP_CODIGO']]
datos_candidatos_2.shape
```

```python
datos_alcaldes = pd.merge(datos_alcaldes, datos_candidatos_2, on="CANDIDATO_CODIGO")
```

```python
datos_alcaldes
```

```python
organizaciones_polticas = pd.read_spss("data/organizacionespolticas2019.sav")
```

```python
organizaciones_polticas.head()
```

```python
organizaciones_polticas_2 = organizaciones_polticas[['OP_CODIGO', 'OP_NOMBRE']] 
```

```python
datos_alcaldes = pd.merge(datos_alcaldes, organizaciones_polticas_2, on="OP_CODIGO")
```

```python
datos_alcaldes.head()
```

```python
datos_loja = datos_alcaldes[datos_alcaldes['CANTON_NOMBRE']=="LOJA"]
datos_loja
```

```python
# datos_loja.groupby("CANDIDATO_NOMBRE")['VOTOS'].sum().sort_values().plot(kind="bar")
datos_loja.groupby("CANDIDATO_NOMBRE")['VOTOS'].sum().sort_values(ascending=False).plot(kind="bar")
```
