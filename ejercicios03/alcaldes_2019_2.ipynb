```python
import pandas as pd
import ipywidgets as widgets
```

```python
datos_alcaldes = pd.read_csv('data/datos_alcaldes_2019_modificado.csv', sep="|")
```

```python
datos_alcaldes
```

```python
canton = widgets.Dropdown(
    options=[('LOJA', "LOJA"), ("CATAMAYO", "CATAMAYO"), ("PINDAL", "PINDAL")],
    value="LOJA",
    description='Cantón:',
)

canton
```

```python
datos_x_canton = datos_alcaldes[datos_alcaldes['CANTON_NOMBRE']==canton.value]
datos_x_canton
```

```python
# datos_loja.groupby("CANDIDATO_NOMBRE")['VOTOS'].sum().sort_values().plot(kind="bar")
datos_x_canton.groupby("CANDIDATO_NOMBRE")['VOTOS'].sum().sort_values(ascending=False).plot(kind="bar")
```
